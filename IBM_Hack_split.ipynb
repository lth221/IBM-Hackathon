```python
## code for file splitting

# make a list that have same landuse code

df_3000 =[] #change
for i in range(0,len(land_csv)):
    if land_geo['features'][i]['properties']['LANDUSE_CODE'] == 3000: #change
        df_new = [land_geo['features'][i]] # make each element in geojson to list
        df_3000 = df_3000 + df_new #change
        df_3000 #change

# convert list to dictionary (because geojson file's format is dictionary)

keys = ['type', 'name', 'crs', 'features']
values = ['FeatureCollection','Statewide_Land_Use_Land_Cover',
          {'type': 'name', 'properties': {'name': 'urn:ogc:def:crs:OGC:1.3:CRS84'}},df_3000] #change
dictionary = dict(zip(keys, values))

# save to geojson file

with open('code3000.geojson', 'a') as f: #change
    json.dump(dictionary, f)
    
## now we get splitted geojson file
```

```python
# scatter plot example

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_us_cities.csv')
df.head()

df['text'] = df['name'] + '<br>Population ' + (df['pop']/1e6).astype(str)+' million'
limits = [(0,2),(3,10),(11,20),(21,50),(50,3000)]
colors = ["royalblue","crimson","lightseagreen","orange","lightgrey"]
cities = []
scale = 5000

fig = go.Figure()

for i in range(len(limits)):
    lim = limits[i]
    df_sub = df[lim[0]:lim[1]]
    fig.add_trace(go.Scattergeo(
        locationmode = 'USA-states',
        lon = df_sub['lon'],
        lat = df_sub['lat'],
        #text = df_sub['text'],
        marker = dict(
            size = df_sub['pop']/scale,
            color = colors[i],
            line_color='rgb(40,40,40)',
            line_width=0.5,
            sizemode = 'area'
        ),
        name = '{0} - {1}'.format(lim[0],lim[1])))

# plot only usa
fig.update_layout(
        title_text = '2014 US city populations<br>(Click legend to toggle traces)',
        showlegend = True,
        geo = dict(
            scope = 'usa',
            landcolor = 'rgb(217, 217, 217)',
        )
    )

fig.show()
```
