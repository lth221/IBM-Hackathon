```python
# make id key

df_geo_id = {}
for feature in df_geo['features']:
    feature['id'] = feature['properties']['OBJECTID']
    df_geo_id[feature['properties']['OGC_NUMBER']] = feature['id']
```

```python
#!pip install geojson_rewind
from geojson_rewind import rewind # for fixing filling error

# The geojson must be unwound before plotting
geo_re = rewind(df_geo,rfc7946=False)

import plotly.express as px

# plotting
fig = px.choropleth(df_csv, geojson=geo_re,
                    locations="OBJECTID", 
                    #color = "blue",
                    #color_continuous_scale="bluered",
                    #scope='usa',
                    title="Verified Waterbodies"
                   )
fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
# Read Algae Bloom Site
df= pd.read_csv("Florida_Algal_Bloom_Site_Visits.csv")
algae = df[df.AlgaeObserved=="Yes"]
algae = algae.drop(columns=["X","Y"])
algae
```

```python
import plotly.graph_objects as go # for scatterplot

fig = px.choropleth(df_csv, geojson=geo_re,
                    locations="OBJECTID", 
                    #scope='usa'
                  )

fig.add_trace(go.Scattergeo(locationmode = 'USA-states',
                            lon = algae['Longitude'],
                            lat = algae['Latitude'],
                            opacity=0.8,
                            marker_color = "Red",
                            showlegend=False))
          
fig.update_geos(showcoastlines = True, 
                resolution = 50,
                fitbounds="locations", visible=False)
fig.show()
```

```python
# enter level2 land use code and run.

code = 1700

###################################################################

land_csv = pd.read_csv("Statewide_Land_Use_Land_Cover.csv")

geo_code = json.load(open("code"+str(code)+".geojson", 'r'))
csv_code = land_csv[land_csv.LEVEL2_LANDUSE_CODE==code]

id_code = {}
for feature in geo_code['features']:
    feature['id'] = feature['properties']['OBJECTID']
    id_code[feature['properties']['OBJECTID']] = feature['id']
    
final_geo_code = rewind(geo_code,rfc7946=False)
fig = px.choropleth(csv_code, geojson=final_geo_code,
                    locations="OBJECTID", 
                     #color = "SHAPELEN",
                    #scope='usa'
                  )

fig.add_trace(go.Scattergeo(
            locationmode = 'USA-states',
               lon = algae['Longitude'],
               lat = algae['Latitude'],
               opacity=0.8,
               marker_color = "Red",
               showlegend=False))

fig.update_geos(showcoastlines = True, 
                resolution = 50,
                fitbounds="locations", visible=False)
fig.show()
```
